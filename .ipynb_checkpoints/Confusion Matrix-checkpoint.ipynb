```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import precision_recall_fscore_support, accuracy_score

def plot_confusion_from_matrix(cm_counts, labels, normalize=True, out_prefix="confmat"):
    """
    cm_counts: 2D numpy array of shape (n_labels, n_labels) with raw counts
    labels: list-like of class labels in the desired order (length n_labels)
    normalize: if True, also plot a row-normalized version (recall)
    Saves two PNGs: <out_prefix>_counts.png and <out_prefix>_normalized.png
    """
    cm_counts = np.asarray(cm_counts)
    n = len(labels)
    if cm_counts.shape != (n, n):
        raise ValueError("cm_counts must be a square matrix with size equal to len(labels)")

    # Row-normalize (recall per true class)
    with np.errstate(all='ignore'):
        cm_norm = cm_counts.astype(float) / cm_counts.sum(axis=1, keepdims=True)
    cm_norm = np.nan_to_num(cm_norm)

    # Create a figure with two heatmaps side-by-side
    fig, axes = plt.subplots(1, 2, figsize=(15, 5))

    sns.heatmap(cm_counts, annot=True, fmt="d", cmap="Blues",
                xticklabels=labels, yticklabels=labels, ax=axes[0])
    axes[0].set_title("Confusion Matrix (counts)")
    axes[0].set_xlabel("Predicted")
    axes[0].set_ylabel("True")

    if normalize:
        sns.heatmap(cm_norm, annot=True, fmt=".2f", cmap="Blues",
                    xticklabels=labels, yticklabels=labels, ax=axes[1])
        axes[1].set_title("Confusion Matrix: GenAI Method")
        axes[1].set_xlabel("Predicted")
        axes[1].set_ylabel("True")
    else:
        sns.heatmap(cm_norm, annot=False, cmap="Blues",
                    xticklabels=labels, yticklabels=labels, ax=axes[1])
        axes[1].set_title("Row-normalized (for reference)")
        axes[1].set_xlabel("Predicted")
        axes[1].set_ylabel("True")

    plt.tight_layout()
    plt_path_counts = f"{out_prefix}_counts.png"
    plt_path_norm = f"{out_prefix}_normalized.png"
    fig.savefig(plt_path_counts, dpi=300)
    fig.savefig(plt_path_norm, dpi=300)
    plt.show()
    print(f"Saved: {plt_path_counts}, {plt_path_norm}")

def per_class_metrics_from_matrix(cm_counts, labels):
    """
    Compute per-class precision, recall, F1 from a given confusion matrix.
    Returns a DataFrame indexed by labels.
    """
    cm = np.asarray(cm_counts)
    # True positives per class are the diagonal
    per_class_tp = np.diag(cm)
    per_class_fp = cm.sum(axis=0) - per_class_tp
    per_class_fn = cm.sum(axis=1) - per_class_tp

    with np.errstate(divide='ignore', invalid='ignore'):
        precision = per_class_tp / (per_class_tp + per_class_fp)
        recall = per_class_tp / (per_class_tp + per_class_fn)
        f1 = 2 * precision * recall / (precision + recall)

    # replace NaN with 0 for any class with zero denominator
    precision = np.nan_to_num(precision)
    recall = np.nan_to_num(recall)
    f1 = np.nan_to_num(f1)

    df = pd.DataFrame({
        "label": labels,
        "precision": precision,
        "recall": recall,
        "f1": f1,
        "support": cm.sum(axis=1)
    }).set_index("label")

    acc = cm_counts.trace() / cm_counts.sum()
    print(f"Overall accuracy (from matrix): {acc:.4f}")
    print("\nPer-class metrics:")
    print(df)
    return df

# Example usage
if __name__ == "__main__":
    # Replace with your actual 6x6 counts matrix in the correct label order
    labels = ['Other', 'Intro','Technology', 'Compliance', 'Economics', 'Legal'] 

    # Example placeholder counts (should be replaced with your real numbers)
    cm_counts = [
        [3,1,3,0,1,4],
        [0,7,0,0,2,2],
        [5,1,12,1,0,1],
        [0,0,3,13,0,1],
        [0,3,0,0,18,0],
        [7,3,2,1,2,58],
    ]

    cm_counts = np.array(cm_counts)

    # Plot two heatmaps (counts + normalized)
    plot_confusion_from_matrix(cm_counts, labels, normalize=True, out_prefix="confusion_6cat")

    # Print per-class metrics
    per_class_metrics_from_matrix(cm_counts, labels)
```
